```python
# Поиск удаленных ребер транзакций между юридическими лицами России 
# (пример: поиск транзакций между организациями в других банках).

# Загрузка библиотек
import numpy as np # для работы с массивами
import pandas as pd # для работы с DataFrame 
from sklearn import datasets # для импорта данных
import tqdm
import catboost as cat
from catboost import CatBoostClassifier

import seaborn as sns # библиотека для визуализации статистических данных
import matplotlib.pyplot as plt # для построения графиков

%matplotlib inline
```

```python
ids = pd.read_csv('ids.csv')
vertices = pd.read_csv('vertices.csv')
vertices['main_okved'] = vertices['main_okved'].astype(str)
#df.info()
vertices
```

```python

idc = list(ids.id)
result = pd.DataFrame (columns=['id_1', 'id_2', 'preds'])

for i in idc:  # tqdm.tqdm(ids.id):
    # соберем датасет из всех возможных вершин
    # вершины имеющие в исходных данных ребро с i обозначим 1, остальные 0
    # учтем то, что вершина i может быть как среди id_1, так и среди id_2
    
    df1 = edges[edges['id_1'] == i].reset_index()
    df2 = edges[edges['id_2'] == i].reset_index()
    df1 = df1[['id_2', 'id_1']].rename(columns={'id_1':'id_2', 'id_2':'id_1'})
    df2= df2[['id_1', 'id_2']]
    df = pd.concat([df1, df2], ignore_index=False)
    df['target'] = 1

    df=df.rename(columns={'id_1':'id'}, inplace=False)
    df = vertices.merge(df,on='id',how='left')
    df['target']=df['target'].fillna(0)
    df['id_2'] = i     # подставляем значение id_2  и для каждой пары id1-id2 далее предсказываем вероятность 
    
    X = df[['main_okved', 'region_code', 'company_type','id_2']]
    y = df['target']
 

    model = CatBoostClassifier(iterations=1000, verbose=False)  # выявлено, чем больше итераций, тем точнее предсказание
    cat_features = [0,1,2,3] # все признаки категориальные
    
    model.fit(X, y, cat_features)

    preds = model.predict_proba(X)
    display(preds)
    df['preds'] = preds[:,1]
  
    
    # соберем общий датасет по всем id2 с вероятностью более 0.02 по каждому случаю
    res_end = df[(df['target'] != 1) & (df['preds']>0.02)].sort_values(by='preds', ascending=False).reset_index()[['id', 'id_2','preds']]
    res_end.columns = ['id_1', 'id_2','preds']
        
    result = result.append(res_end, ignore_index=True, sort=False)
```

```python
result.drop_duplicates()
```

```python
# В итоговый датасет отберем необходимое нам количество записей с наибольшей вероятностью.

res_all=result.sort_values(by='preds', ascending=False).iloc[:100000].reset_index()[['id_1', 'id_2']]
res_all.columns = ['id_1', 'id_2']
```

```python
# Сохраним результаты в файл/

res_all.to_csv('submission_t4.csv', index = False)
```
